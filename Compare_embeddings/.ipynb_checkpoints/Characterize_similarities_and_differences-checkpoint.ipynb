```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

# Mistral vs GPT

```python
# Load embeddings files

# Input the path to your saved embeddings
path1 = '/Users/ellenbowen/Desktop/Michigan/Siwo_lab/'
file1 = str(path1) + 'phecode_mistral_embedding.pkl'

df_mistral = pd.read_pickle(file1)


path2 = '/Users/ellenbowen/Desktop/Michigan/Siwo_lab/'
file2 = str(path2) + 'phecode_gpt_embedding.pkl'

df_gpt = pd.read_pickle(file2)

path3 = '/Users/ellenbowen/Desktop/Michigan/Siwo_lab/'
file3 = str(path3) + 'phecode_voyage_embedding.pkl'

df_voyage = pd.read_pickle(file3)
```

```python
# Function to calculate cosine similarities for a given set of embeddings
def calculate_cosine_similarity(df, embedding_column='embeddings'):
    embeddings = np.stack(df[embedding_column].values)
    similarity_matrix = cosine_similarity(embeddings)
    return similarity_matrix

# Calculate cosine similarities for each LLM
mistral_similarities = calculate_cosine_similarity(df_mistral)
gpt_similarities = calculate_cosine_similarity(df_gpt)
voyage_similarities = calculate_cosine_similarity(df_voyage)

# Get the pairwise correlations (cosine similarities) between diseases for both models
mistral_flat = mistral_similarities[np.triu_indices_from(mistral_similarities, k=1)]
gpt_flat = gpt_similarities[np.triu_indices_from(gpt_similarities, k=1)]
voyage_flat = voyage_similarities[np.triu_indices_from(voyage_similarities, k=1)]

```

```python
# Create a scatter plot of the correlations
plt.figure(figsize=(6, 6))
plt.scatter(mistral_flat, gpt_flat, alpha=0.5)
plt.title('Correlation Comparison: Mistral vs GPT')
plt.xlabel('Mistral Cosine Similarities')
plt.ylabel('GPT Cosine Similarities')
plt.grid(True)
plt.show()
```

# Mistral vs Voyage

```python
# Create a scatter plot of the correlations
plt.figure(figsize=(6, 6))
plt.scatter(mistral_flat, voyage_flat, alpha=0.5)
plt.title('Correlation Comparison: Mistral vs Voyage')
plt.xlabel('Mistral Cosine Similarities')
plt.ylabel('Voyage Cosine Similarities')
plt.grid(True)
plt.show()
```
